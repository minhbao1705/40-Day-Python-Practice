```python
from newspaper import Article

# To mt i tng Article tu URL ca bi bo
article = Article('https://vnexpress.net/thoi-tiet-mien-bac-mien-trung-mien-nam-ngay-14-3-4518045.html')

# Ti bi bo
article.download()
article.parse()

# In bi bo
print(article.text)

# Ly ton b nh trong bi bo
print(article.images)
```

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')

data = "Ti thch hc AI v Ton"
# Bc 1: Tokenization data
tokenization = word_tokenize(data)
# Bc 2: gi th vin Pos tagging
result = nltk.pos_tag(tokenization)
print(result)
```
