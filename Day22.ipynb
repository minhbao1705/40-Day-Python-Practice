```python
import random

def random_number_with_codition(total):
    random.seed(0)
    sum = 0
    count=0
    
    
    while(sum!=total):
        a = random.randint(1, 20)
        b = random.randint(1, 20)
        sum = a+b
        count+=1
    
    return count

print(f"Test case 1: Random_number_with_codition(40): ", random_number_with_codition(40))
print(f"Test case 2: Random_number_with_codition(20): ", random_number_with_codition(20))
print(f"Test case 3: Random_number_with_codition(35): ", random_number_with_codition(35))
    
```

```python
def find_diivisible_number(a):
    x = 101
    while(True):
        if x%a==0:
            break
        elif x%a!=0:
            x+=1
    return x

print("Test case 1: find_divisible_number(5): ", find_diivisible_number(5))
print("Test case 2: find_divisible_number(17): ", find_diivisible_number(17))
```

```python
def f(x, a):
    return x*x-a

def fn(x):
    return 2*x


def find_squared_root(a):
    EPSILON=0.001
    xn=a
    xn1 = xn - f(a, a)/fn(a)
    while(abs(xn1-xn)>EPSILON):
        xn = xn1
        xn1 = xn - f(xn, a)/fn(xn)
    return xn1

print("Test case 1: find_squared_root(2): ", find_squared_root(2))
print("Test case 2: find_squared_root(3): ", find_squared_root(3))
```
