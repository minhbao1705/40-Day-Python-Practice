```python
# Câu 1
lst_data = [2, 4, 6, 8, 10, 12]
print(lst_data)
```

```python
# Câu 2
for i in lst_data:
    if i%3==0:
        lst_data.remove(i)
print(lst_data)
```

```python
# Câu 3
# Them vào cuối 1 chuỗi các số từ 1 đến 3
for i in range(1, 4):
    lst_data.append(i)
print(lst_data)
```

```python
# Them vào vị trí index thứ 3 chuỗi các số từ 6 đến 8
position = 3
value = 6
for i in range(3):
    lst_data.insert(position, value)
    position += 1
    value += 1
print(lst_data)
```

```python
# Câu 4
for i, item in enumerate(lst_data):
    if lst_data[i]%2==0 or lst_data[i]%5==0:
        lst_data[i] = 0
print(lst_data)
```
