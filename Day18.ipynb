```python
stop_words = ["I", "love", "and", "to"]
input = "I love AI and listen to music"
lst_data = input.split()
lst=[]
for i in lst_data:
    if i not in stop_words:
        lst.append(i)

words = [x for x in lst_data if x not in stop_words]
print("For: ", lst)
print("Comprehension: ",words)
```

```python
import math
my_tuple1 = (2, 3)
my_tuple2 = (3, 6)

sum1=0
sum2=0
product1=1
product2=1
for i in my_tuple1:
    sum1 += i
    product1 = product1*i
result_vector1 = (sum1, product1)

for i in my_tuple2:
    sum2 += i
    product2 = product2*i
result_vector2 = (sum2, product2)

print(f"resut_vector1: {result_vector1}, result_vector2: {result_vector2}")

```

```python
# Câu 3
square_sum = 0
for x, y in zip(my_tuple1, my_tuple2):
    square_sum += (x-y)**2
    distance = math.sqrt(square_sum)

print(distance)
```
