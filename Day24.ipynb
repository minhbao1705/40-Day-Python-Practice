```python
# Question 5
def my_function(signal):
    var = True
    while var:
        var = False
        for i in range(len(signal) - 1):
            if signal[i] > signal[i + 1]:
                signal[i], signal[i + 1] = signal[i + 1], signal[i]
                var = True

my_signal = [1, 0, 0, 1, 1, 1]
my_function(my_signal)
print(my_signal)
```

```python
# Question 18
def my_function(data, max, min):
    result = []
    for i in data:
        if i < min:
            result.append(min)
        elif i > max:
            result.append(max)
        else: 
            result.append(i)
    return result

my_list = [1, 2, 3, 4, 5, 6, 7]
max = 5
min = 2
print(my_function(my_list, max, min))
```
