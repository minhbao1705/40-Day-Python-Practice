### To Bag of words

```python
corpus = ["Ti thch mn Ton", "Ti thch AI", "Ti thch m nhc"]

data = ' '.join(corpus)
data
```

```python
distinct_words = list(set(data.split()))
length = len(distinct_words)

distinct_words
```

```python
for i, word in enumerate(distinct_words):
    print(i, word)
```

### Thc hin vectorization

```python
sample = "Ti thch AI thch Ton"
sample_vec = [0]*length
sample_words = sample.split()
for word in sample_words:
    sample_vec[distinct_words.index(word)] +=1
print(sample_vec)
```
