```python
# Th vin Translate
from translate import Translator

translator = Translator(to_lang="vi")
translation = translator.translate("Don't cry because it's over, smile because it happened.")
print(translation)
```

```python
# Th vin Googletrans: Dch t ting Vit sang ting Php
from googletrans import Translator

# define a translate object
translate = Translator()

# Translate some text
result = translate.translate('Hm nay thi tit khng tt v ma nhiu', dest='fr')

print(result)
print(result.text)
```

### 2. Bi tp

```python
sentence = "Cng c Suno AI nhanh chng c s ch  t ngi dng khi c th to bi ht ch vi vi cu lnh. Phin bn mi nht V3 Alpha mi c gii thiu cui thng 2, c bn min ph vi 10 bi ht mi ngy."
```

```python
# Th vin Translate
from translate import Translator

translator = Translator(from_lang="vi", to_lang="en")
translation = translator.translate(sentence)
print(translation)

translator = Translator(from_lang="vi", to_lang="ja")
translation = translator.translate(sentence)
print(translation)
```

```python
# Th vin Googletrans
from googletrans import Translator

# define a translate object
translate = Translator()

# Translate some text
result = translate.translate(sentence, dest='en')
print(result)
print(result.text)

# Translate some text
result = translate.translate(sentence, dest='ja')
print(result)
print(result.text)
```
