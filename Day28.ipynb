```python
# Thư viện Translate
from translate import Translator

translator = Translator(to_lang="vi")
translation = translator.translate("Don't cry because it's over, smile because it happened.")
print(translation)
```

```python
# Thư viện Googletrans: Dịch từ tiếng Việt sang tiếng Pháp
from googletrans import Translator

# define a translate object
translate = Translator()

# Translate some text
result = translate.translate('Hôm nay thời tiết không tốt và mưa nhiều', dest='fr')

print(result)
print(result.text)
```

### 2. Bài tập

```python
sentence = "Công cụ Suno AI nhanh chóng được sự chú ý từ người dùng khi có thể tạo bài hát chỉ với vài câu lệnh. Phiên bản mới nhất V3 Alpha mới được giới thiệu cuối tháng 2, có bản miễn phí với 10 bài hát mỗi ngày."
```

```python
# Thư viện Translate
from translate import Translator

translator = Translator(from_lang="vi", to_lang="en")
translation = translator.translate(sentence)
print(translation)

translator = Translator(from_lang="vi", to_lang="ja")
translation = translator.translate(sentence)
print(translation)
```

```python
# Thư viện Googletrans
from googletrans import Translator

# define a translate object
translate = Translator()

# Translate some text
result = translate.translate(sentence, dest='en')
print(result)
print(result.text)

# Translate some text
result = translate.translate(sentence, dest='ja')
print(result)
print(result.text)
```
