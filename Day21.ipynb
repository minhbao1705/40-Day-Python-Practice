```python
import operator
words = data.split()
def count_word(words, distinct_words):
    count = {}
    for small_word in distinct_words:
        for word in words:
            if small_word in word:
                if small_word not in count:
                    count[small_word] = 1
                else:
                    count[small_word] += 1
    max_value = 0
    max_key = None
    for key, value in count.items():
        if value > max_value:
            max_value = value
            max_key = key

    min_value = 1
    min_key = None
    for key, value in count.items():
        if value <= min_value:
            min_value = value
            min_key = key

    return count, max_key, min_key

res, max, min = count_word(words, distinct_words)
# max_key = int(max(res.values()))
# max_value = res[max_key]

for key, value in res.items():
    print(f"\"{key}\" in data is {value}")
print(max, "is most frequent word")
print(min, "is the least common word")
```
