```python
lst_data = [
    [1, 2, 3],
    [4, 5, 6], 
    [7, 8, 9]
]

lst_sub_data = []
for lst in lst_data:
    sub = [val for idx, val in enumerate(lst) if idx == 0 or idx == 2]
    lst_sub_data.append(sub)

print(lst_sub_data)
```

```python
mat_a = [
     [1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]
     ]

mat_b = [
     [2, 4, 6],
     [1, 3, 5],
     [1, 0, 1]
     ]
sum = []
for i in range(len(mat_a)):
    total_matrix=[]
    for j in range(len(mat_a)):
        total = mat_a[i][j] + mat_b[i][j]
        total_matrix.append(total)
    sum.append(total_matrix)

print("Tng bng: ",sum)

sub = []
for i in range(len(mat_a)):
    total_matrix=[]
    for j in range(len(mat_a)):
        total = mat_a[i][j] + mat_b[i][j]
        total_matrix.append(total)
    sub.append(total_matrix)

print("Hiu bng: ", sub)

dot = []
for row in range(len(mat_a)):
    total_matrix=[]
    for col in range(len(mat_b)):
        sum = 0
        for i in range(len(mat_b)):
            sum+= mat_a[row][i] * mat_b[i][col]
        total_matrix.append(sum)
    dot.append(total_matrix)

print("dot product: ", dot)
```
